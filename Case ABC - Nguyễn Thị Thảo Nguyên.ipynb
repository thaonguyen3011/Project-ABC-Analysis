```python

import pandas as pd
df = pd.read_csv('E:/Data analyst for python/INPUT_sales.csv',usecols=['StockCode','Quantity','Description','Price'])
df['Revenue'] = df['Quantity']*df['Price']
df = df[df['Price']>=0].groupby('StockCode').agg({'Quantity':'sum','Revenue':'sum'}).sort_values(by='Revenue',ascending=False).reset_index()
#Những mã hàng nào thuộc nhóm nào 
df['CumSumRevenue'] = df['Revenue'].cumsum()
df['%CumSumRevenue'] =  df['CumSumRevenue']/sum(df['Revenue'])*100
df['Class'] = pd.cut(x = df['%CumSumRevenue'], bins=[1,80,95,100],labels=['A','B','C'])
df 
#Số lượng mã hàng theo từng nhóm
df1 = df.groupby('Class').agg({'StockCode':'count','Revenue':'sum'}).rename(columns ={'StockCode':'number_of_sku'}).reset_index()
df1['sales_percentage'] =  (df1['Revenue']/sum(df1['Revenue'])).round(2)
df1['sku_percentage'] = (df1['number_of_sku']/sum(df1['number_of_sku'])).round(2)
df1
```
